```python
import pandas as pd
import seaborn as sns
import requests
import json
import datetime
import sqlite3
import sqlalchemy
```

```python
engine= sqlalchemy.create_engine('sqlite:///database.db') #Create a database and connection
link= "https://www.gov.uk/bank-holidays.json"
fetch=requests.get(link)
fetch #to check the status 200
json_result= fetch.json()
```

```python
json_result
```

```python
df2=pd.json_normalize([json_result])
df2
```

```python
json_list=[] #unfollding the dictionary

for key in json_result.keys():
    df= pd.DataFrame(json_result[key]['events'])
    df['division']= json_result[key]['division']
    json_list.append(df)
    
```

```python
json_list
```

```python
data= pd.concat(json_list).reset_index(drop=True)
data
```

```python
#Change the format of the data
data[('date')] = pd.to_datetime(data[('date')]).dt.strftime('%d-%m-%Y') 
data 
```

```python
#converting the dates in to days
date_list=[]

for list_var in data.date:
    day, month, year = (int(list_var) for list_var in list_var.split('-'))    
    ans = datetime.date(year,month,day)
    list_var2=ans.strftime("%A")
    date_list.append(list_var2)
```

```python
data['week_days']=pd.DataFrame(date_list)
```

```python
data.head(50)
```

```python
#Overwrite the data into by using replace command
data.to_sql('Holidays',engine, if_exists='replace', index=False)
```
